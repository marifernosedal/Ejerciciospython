```python
def acceso_boveda(edad_juan, edad_pedro, num_nietos):
    condicion_a = edad_juan > 35 or edad_pedro < 50
    condicion_b = (edad_juan + edad_pedro) % 2 == 0
    condicion_c = num_nietos >= 1
    return "SI" if condicion_a and condicion_b and condicion_c else "NO"

juan = int(input("Edad de Juan: "))
pedro = int(input("Edad de Pedro: "))
nietos = int(input("Número de nietos de Manuel: "))
print(f"¿Tienen acceso?: {acceso_boveda(juan, pedro, nietos)}")

```

```python
#Ejercicio 2
# --- Entrada: HH:MM de Ciudad de México ---
entrada = input("Ingresa la hora en Ciudad de México (HH:MM o HH, 24h): ").strip()

# Validación básica
try:
    parts = entrada.split(':')
    if len(parts) == 2:
        hh, mm = map(int, parts)
    elif len(parts) == 1:
        hh = int(parts[0])
        mm = 0 # Assume minutes are 0 if only hour is given
    else:
        raise ValueError("Formato de hora inválido")

    assert 0 <= hh < 24 and 0 <= mm < 60
except ValueError:
    print("Error: usa formato HH:MM válido (p.ej. 13:30) o solo HH (p.ej. 9).")
    raise SystemExit
except AssertionError:
    print("Error: La hora u minutos están fuera de rango (0-23 para horas, 0-59 para minutos).")
    raise SystemExit

# Convertimos a minutos desde medianoche
cdmx_min = hh*60 + mm

# --- Diferencias (en minutos) respecto a Ciudad de México ---
# Nota: simplificado SIN horario de verano (lo que piden en la tarea).
dif = {
    "Tokio, Japón":               15*60,
    "Pekín, China":               14*60,
    "Dusambé, Tayikistán":        11*60,
    "Perth, Australia":           15*60,
    "Katmandú, Nepal":            11*60 + 45,   # 11h 45min
    "Vilna, Lituania":             8*60,
    "Freetown, Sierra Leona":      6*60,
    "Río de Janeiro, Brasil":      3*60,
    "Tegucigalpa, Honduras":      -1*60,
    "La Paz, Baja California Sur, México": -1*60
}

# --- Orden deseado (lista) ---
orden = [
    "Tokio, Japón", "Pekín, China", "Dusambé, Tayikistán", "Perth, Australia",
    "Katmandú, Nepal", "Vilna, Lituania", "Freetown, Sierra Leona",
    "Río de Janeiro, Brasil", "Tegucigalpa, Honduras",
    "La Paz, Baja California Sur, México"
]

# --- Función lambda para sumar offset y hacer la vuelta de 24h ---
calc = lambda base_min, delta_min: (base_min + delta_min) % (24*60)

# --- Helper para formatear HH:MM con cero a la izquierda ---
def fmt(min_tot):
    h = min_tot // 60
    m = min_tot % 60
    return f"{h:02d}:{m:02d}"

print("\nHora en las siguientes ciudades:")
for ciudad in orden:
    m = calc(cdmx_min, dif[ciudad])
    print(f"{ciudad}: {fmt(m)}")
```

```python
# Ejercicio 3: Número primo
# Determina si un número es primo, validando que sea mayor que 1.

def es_primo():
    while True:
        n = int(input("Ingresa un número entero: "))
        if n <= 1:
            print("Error en el número, inténtalo de nuevo.")
        else:
            break
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            print("False")
            return
    print("True")

es_primo()
```

```python
# Ejercicio 4: Números mayores a 100
# Evalúa el número según sus últimas cifras y divisiones.

def analizar_numero():
    while True:
        n = int(input("Ingresa un número entero mayor que 100: "))
        if n <= 100:
            print("Error en el número, inténtalo de nuevo.")
        else:
            break

    ultimo = n % 10
    if ultimo in [4, 7, 9]:
        print(f"¿Es divisible por 3?: {'Sí' if n % 3 == 0 else 'No'}")
    elif ultimo in [1, 5]:
        print(f"Resultado de N/10: {n // 10}")
    else:
        print(f"Residuo de N/3: {n % 3}")

analizar_numero()
```

```python
# Ejercicio 5: MCD y MCM
# Calcula el Máximo Común Divisor y el Mínimo Común Múltiplo de dos números.

import math

def mcd_mcm(a, b):
    resultados = {
        "MCD": math.gcd(a, b),
        "MCM": abs(a*b) // math.gcd(a, b)
    }
    return resultados

a = int(input("Ingresa a: "))
b = int(input("Ingresa b: "))
print(mcd_mcm(a, b))
```

```python
# Ejercicio 7: Fórmula de Haversine
# Calcula la distancia (en km) entre dos puntos geográficos.

import math

def haversine(lat1, lon1, lat2, lon2):
    R = 6378.1
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    dlat, dlon = lat2 - lat1, lon2 - lon1
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2)*math.sin(dlon/2)**2
    return 2 * R * math.asin(math.sqrt(a))

print(f"París–Roma: {haversine(48.9, 2.4, 41.9, 12.5):.0f} km")
print(f"CDMX–Nueva York: {haversine(19.43, -99.13, 40.71, -74.00):.0f} km")
print(f"CDMX–Tokio: {haversine(19.43, -99.13, 35.68, 139.69):.0f} km")
print(f"CDMX–Buenos Aires: {haversine(19.43, -99.13, -34.61, -58.38):.0f} km")
```

```python
# Ejercicio 8: Serie de Fibonacci
# Genera la serie de Fibonacci con validación de entrada.

def fibonacci(n):
    # Validación simple
    if n <= 0:
        return "Error en el número"   # o return [] si prefieres lista vacía

    a, b = 0, 1
    serie = []
    for _ in range(n):
        serie.append(a)
        a, b = b, a + b
    return serie

# --- con input del usuario ---
try:
    n = int(input("¿Cuántos términos de Fibonacci quieres? "))
    resultado = fibonacci(n)
    print(resultado)
except ValueError:
    print("Entrada inválida: debe ser un entero.")
```

```python
# Ejercicio 9: Catálogo de sismos
# Analiza listas de eventos sísmicos, convierte datos y muestra resultados ordenados.

Fechas = ["2025-03-01","2025-03-02","2025-03-03","2025-03-04","2025-03-05",
          "2025-03-06","2025-03-07","2025-03-08","2025-03-09","2025-03-10"]

Zonas = ["guerrero","oaxaca","chiapas","michoacán","guerrero",
         "oaxaca","chiapas","guerrero","oaxaca","chiapas"]

magnitudes = ["3.8","4.2","5.0","2.9","4.8","3.5","4.1","5.3","4.0","3.0"]
prof_km = [15, 80, 12, 5, 35, 120, 60, 25, 10, 90]

# 1. Convertir magnitudes a float
magnitudes = list(map(float, magnitudes))

# 2. Capitalizar nombres con lambda
Zonas = list(map(lambda z: z.capitalize(), Zonas))

# 3. Clasificar intensidad
intensidad = []
for m in magnitudes:
    if m < 3.5: intensidad.append("Muy leve")
    elif m < 4.5: intensidad.append("Leve")
    elif m < 5.5: intensidad.append("Moderado")
    else: intensidad.append("Fuerte")

# 4. Conteo por zona
conteo_por_zona = {}
for z in Zonas:
    conteo_por_zona[z] = conteo_por_zona.get(z, 0) + 1

# 5. Ordenar por magnitud descendente y mostrar top 3
eventos = list(zip(Fechas, Zonas, magnitudes, prof_km))
eventos.sort(key=lambda x: x[2], reverse=True)

print("Top 3 sismos más grandes:")
for e in eventos[:3]:
    print(f"{e[0]} | {e[1]} | {e[2]} | {e[3]} km")

print("\nConteo por zona:", conteo_por_zona)
```

```python
# Ejercicio 10: Operaciones vectoriales
# Calcula productos escalares y vectoriales entre tres vectores y analiza si son paralelos o perpendiculares.

import numpy as np

a = np.array([2, -1, 4])
b = np.array([4, -2, 8])
c = np.array([-5, 2, 1])

def vectores(v1, v2):
    escalar = np.dot(v1, v2)
    vectorial = np.cross(v1, v2)
    return escalar, vectorial

pares = [("a", a, "b", b), ("b", b, "c", c), ("a", a, "c", c)]
for n1, v1, n2, v2 in pares:
    esc, vec = vectores(v1, v2)
    print(f"{n1}{n2}: Producto escalar = {esc}, Producto vectorial = {vec}")

def relacion(v1, v2):
    esc = np.dot(v1, v2)
    cruz = np.cross(v1, v2)
    if np.allclose(cruz, [0,0,0]):
        print("Paralelos")
    elif esc == 0:
        print("Perpendiculares")
    else:
        print("Ninguna relación")

print("\nRelaciones:")
relacion(a,b)
relacion(b,c)
relacion(a,c)
```
