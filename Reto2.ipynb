**1.- La secuencia del cortador de pizza perezoso ( The Lazy Caterer's sequence).**

Se trata un problema clásico de combinatoria y geometría que responde a la siguiente pregunta, ¿cuál es el número máximo de piezas en que se puede dividir una pizza utilizando *n* cortes rectos?
Si tienes una pizza y haces cortes rectos, no necesariamente pasando por el centro y con la posibilidad de cruzarse en cualquier punto, ¿cuál es la mayor cantidad de rebanadas que puedes obtener?. Sabemos que

f(0)=1

f(1)=2

f(2)=4

f(3)=7

Con base en esta información, se puede utilizar la siguiente fórmula de recursión para *n* cortes.

\begin{equation}
f(n)=f(n−1)+n
\end{equation}

Genere un programa en Python que muestre la secuencia de n cortes. Proponga un valor apropiado de cortes y justifique porqué escogió dicho valor.

```python
###f(0)=1
###f(n)=f(n-1)+n
def cortador_perezoso(n_cortes):
  valores = [1] # f(0)=1
  for i in range(1, n_cortes + 1):
    siguiente = valores[-1] + i # f(i)=f(i-1)+i
    valores.append(siguiente)
  return valores

n_input = input("¿Cuántos cortes n quieres?")
n = int(n_input)

secuencia = cortador_perezoso(n)
print("\nSecuencia del cortador de pizza perezoso:")
for k in range(n + 1):
  print(f"({k})={secuencia[k]}")
print(f"\nCon {n} cortes se pueden obtener como máximo {secuencia[-1]} rebanadas de pizza.")

#Elegí n=8 cortes porque, normalmente las pizzas tienen 8 rebanadas. Además creo que con n=8 se puede apreciar muy bien que la secuencia crece de forma acelerada y no lineal,
#ya que cada corte adicional aumenta varias piezas, evidenciando el comportamiento  propio de la fórmula recursiva utilizada.
```

**2.- Serie de Madhava.**

Utilice un ciclo for para estimar el valor de pi ($\pi$) a partir de los primeros 20 términos de la Serie de Madhava, dada por la siguiente expresión:

\begin{equation}
\pi \approx \sqrt{12}\left(
\sum_{k=0}^{19} \frac{(-1)^k}{(2k+1)3^k}
\right)
\end{equation}

```python
import math
suma=0
#Ciclo for usando los primeros 20 términos
for k in range(20):
  termino=((-1)**k)/((2*k +1)*(3**k))
  suma+=termino
##Fórmula para aproximar pi
pi_aprox=math.sqrt(12)*suma
print("Aproximación de pi usando los primeros 20 términos de la Serie de Madhava :", pi_aprox)
math.pi
print("Valor real de pi :", math.pi)
```
