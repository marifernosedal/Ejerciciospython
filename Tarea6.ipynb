```python
#7.
A = np.array([[11, 7, -1],
              [-3, 5, 2],
              [4, 2, -2]])

B = np.array([[1, 1, 1],
              [10, 3, -5],
              [-6, -10, -3]])

print("2B + 3A:")
print(2*B + 3*A)

print("(B + 2A)^2:")
print((B + 2*A) @ (B + 2*A))

print("A^2 + 3B:")
print(A @ A + 3*B)
```

```python
#8.
import numpy as np

def procesar_matriz():
    while True:
        try:
            n = int(input("Ingresa n (3 a 5): "))
            m = int(input("Ingresa m (3 a 5): "))
        except:
            print("Entrada inválida\n")
            continue

        if n != m:
            print("La matriz no es cuadrada. Intenta otra vez.\n")
        else:
            break

    # Crear matriz
    T = np.random.randint(-20, 20, (n, m))
    print("\nMatriz T:\n", T)

    # Primera fila invertida
    print("\nPrimera fila invertida:", T[0, ::-1])

    # Primera columna invertida
    print("Primera columna invertida:", T[::-1, 0])

    # Primer y último elemento
    print("\nPrimer elemento:", T[0, 0])
    print("Último elemento:", T[-1, -1])

    # Valores
    print("\nMáximo:", T.max())
    print("Mínimo:", T.min())

    diag = np.diag(T)
    print("Promedio diagonal:", np.mean(diag))
    print("Traza:", np.trace(T))

    # Diagonales
    print("\nDiagonal principal:", diag)
    antidiagonal = np.diag(np.fliplr(T))
    print("Antidiagonal:", antidiagonal)

    # Sumas
    print("\nSuma elementos pares:", np.sum(T[T % 2 == 0]))
    print("Suma elementos impares:", np.sum(T[T % 2 != 0]))

    # Promedios
    print("\nPromedio filas:", np.mean(T, axis=1))
    print("Promedio columnas:", np.mean(T, axis=0))

    print("\nPromedio antidiagonal:", np.mean(antidiagonal))

# LLAMADA (IMPORTANTE):
procesar_matriz()
```

```python
import numpy as np
import matplotlib.pyplot as plt

#1.
x = np.linspace(-5, 5, 500)

C_pos = [1, 3, 5]
C_zero = 0
C_neg = [-1, -3, -5]

plt.figure(figsize=(8,6))

# C > 0
for C in C_pos:
    y = (x - 1) + C*np.exp(-x)
    plt.plot(x, y, label=f"C = {C} (C>0)")

# C = 0
plt.plot(x, x - 1, linewidth=3, label="C = 0")

# C < 0
for C in C_neg:
    y = (x - 1) + C*np.exp(-x)
    plt.plot(x, y, label=f"C = {C} (C<0)")

plt.title("Familia de soluciones de y = x - 1 + C e^{-x}")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
#2.
x = np.linspace(-10, 10, 2000)
y = (3*x + 7) / (x**2 - 4)

plt.figure(figsize=(8,6))

plt.plot(x, y, label="f(x) = (3x+7)/(x²−4)")

# Asíntotas verticales
plt.axvline(2, linestyle="--", label="x = 2")
plt.axvline(-2, linestyle="--", label="x = -2")

# Asíntota horizontal
plt.axhline(0, linestyle="--", label="y = 0")

plt.ylim(-20, 20)
plt.legend()
plt.grid()
plt.show()
```
